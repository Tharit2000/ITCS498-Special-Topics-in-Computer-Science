# Data Preparation

### Use 'F' and 'M' instead of '0' and '1'

```python
df['sex'].replace(['0','1'],['F','M'], inplace=True)
df
```

### Check if there is a null value

```python
df.isnull().sum()
```

### Replace null with 0 in 'ord3' column

```python
df['ord3'].fillna(0, inplace=True)
df
```

### Drop all rows that include null

```python
df.dropna(inplace=True)
df
```

### Change data in all 'ord' columns to category

```python
df = df.astype({'ord1':'category', 'ord2':'category', 'ord3':'category', 'ord4':'category', 'ord5':'category', 'ord6':'category'})
df.info()
```

### Change data in 'target' and 'sex' columns to category

```python
df = df.astype({'target':'category', 'sex':'category'})
df.info()
```

### Joint plots for two pair of features with the highest correlation score
The figure below shows that positive patients are approximately 50 to 60 years old with the num3 value of 150 to 170.

```python
g = sns.jointplot(x='age', y='num3', data = df[(df.target == 1)], kind='kde')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.fig.suptitle('Distribution of age and num3 for whose test result for a disease is positive')
```

The figure below shows that most positive patients have the num3 value of 160 to 180 and the num4 value of 0.

```python
g = sns.jointplot(x='num3', y='num4', data = df[(df.target == 1)], kind='kde', color='m')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.fig.suptitle('Distribution of num3 and num4 for whose test result for a disease is positive')
```
